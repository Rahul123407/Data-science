```python
columns_to_remove=['date','yr_renovated','street','city','statezip','country']
df = df.drop(columns_to_remove,axis=1)
```

```python
df.head()
```

```python
import scipy.stats as stats
z_scores = stats.zscore(df)
threshold=3
print('size before removing outliners',df.shape)
outliers_df= df[(z_scores>threshold).any(axis=1)]
df=df[(z_scores<=threshold).all(axis=1)]
print('size after removing outliners',df.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
df_scaled =pd.DataFrame(scaler.transform(df),columns=df.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x=df.drop('price',axis=1)
y=df['price']
x_train, x_test, y_train ,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
models=[('RandomForest',RandomForestRegressor()),('LinearRegression',LinearRegression()),('Decision Tree',DecisionTreeRegressor()),('KNN',KNeighborsRegressor())]
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
for name,model in models:
    print(name)
    print()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print("MSE",mean_squared_error(y_test,y_pred))
    print("\n")
    print("MAE",mean_absolute_error(y_test,y_pred))
    print("\n")
    print("RQ2",r2_score(y_test,y_pred))
    print("\n")
```
